```
%%time
response = model.generate_content("Tell me a fun fact about the ocean.",stream=True)
```

```
response.prompt_feedback
```

```
for chunk in response:
  if chunk.candidates:
    print(chunk.text)
    print("_"*80)
  else:
    print("Response was blocked or empty.")
    print("_"*80)
```

```
try:
  response.resolve() # Resolve the streaming response
  if response.candidates:
    print(response.text)
  else:
    print("Response has no candidates (potentially blocked).")
except Exception as e:
  print(f'{type(e).__name__}: {e}')
```
